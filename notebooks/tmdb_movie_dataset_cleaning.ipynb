```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('TMDB_movie_dataset_v11.csv')
df.head()
```

```python
df.status.value_counts()
```

```python
# remove unreleased movies
df = df[df['status'] == 'Released'].reset_index(drop=True)
df
```

```python
# remove movies that were only for adults
df = df[df['adult'] == False].reset_index(drop=True)
df
```

```python
# remove movies with vote count = 0
df = df[df['vote_count'] != 0]
df.reset_index(drop=True, inplace=True)
df
```

```python
# remove movies with no release_date
df = df[df['release_date'].notna()].reset_index(drop=True)
df.isnull().sum()
```

```python
# drop movie data which does not have any overview
df = df[df['overview'].notna()].reset_index(drop=True)
df.isnull().sum()
```

```python
# remove movies with no poster path
df = df[df['poster_path'].notna()].reset_index(drop=True)
df.isnull().sum()
```

```python
# convert release_date to datetime obj
df['release_date'] = pd.to_datetime(df['release_date'])
df.dtypes
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# remove unnecessary cols
df.drop(columns=['id', 'status', 'adult','budget', 'revenue', 'backdrop_path','homepage', 'imdb_id', 'keywords', 'original_title',
                 'tagline', 'production_companies', 'production_countries', 'spoken_languages'], inplace=True)
df.columns
```

```python
df.dtypes
```

```python
# fix movie poster path by providing full url

prefix = "https://image.tmdb.org/t/p/w300"

# Update the poster URL column with the prefix
df['poster_path'] = df['poster_path'].apply(lambda x: f"{prefix}{x}")
df.head()
```

```python
df.original_language.value_counts()
```

```python
!pip install langcodes
import langcodes
```

```python
# convert the language codes in original_language column (ISO 639 code) to English name of language
def convert_language_code(code):
  try:
    language = langcodes.Language.get(code).display_name()
  except:
    language = 'Unknown'
  return language

df['original_language'] = df['original_language'].apply(convert_language_code)
df['original_language'].value_counts()
```

```python
df.isnull().sum()
```

```python
# fill those rows with N/A genres with "Unknown"
df['genres'] = df['genres'].fillna('Unknown')
df.isnull().sum()
```

```python
len(df)
```

```python
# append year to movies
df['title'] = df.apply(lambda row: row['title'] + ' (' + str(row['release_date'].year) + ')', axis=1)
df.head()
```

```python
# save the csv
df.to_csv('tmdb_movies.csv', index=False)
```
